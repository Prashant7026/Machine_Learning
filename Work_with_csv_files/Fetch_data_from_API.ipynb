```python
import pandas as pd
import requests
```

When you run below shell then python hit the http request to api by using request library and api give response in form of data which we can see.

```python
# Code to get page 1 from an API ---> In last of URL we see page=1
# Response [200] successfully runs
# Response [404] unavaliable
# Response [500] server down
response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1')
response
```

```python
# Convert incoming data to json format
response.json()
```

```python
# Our use is only result
response.json()['results']
# Now we can see result is in list
```

```python
# Convert list into dataframe
# This is page 1 only
pd.DataFrame(response.json()['results'])
```

```python
# Remove unwanted columns
# Here we successfully got 1 page rest all pages(551) we got in next steps
temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
temp_df
```

# After getting response by API url all steps we also do in just one line --- >

``pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]``

In next steps we get all 551 pages

```python
# Create empty dataframe
df = pd.DataFrame()
df
```

```python
# We perform temp_df step in loop to get 552 pages
# ('_______page={}'.format(i)) is used for getting all 554 pages one by one and then convert it into json and then convert it into dataframe
# After that append every dataframe one by one in df variable and when loop is end then we got full data in df variable
for i in range(1,553):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    df = df.append(temp_df, ignore_index = True)       # Ignore index create their own indexing and give a correct index
```
