# Univariate Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sibsp --> sibling spouse
# parch --> Parent child
# embarked --> Where the traveler mounted from

df = pd.read_csv(r'E:\CETPA\DA\titanic_data.csv')
df.sample(5)
```

## 1. Categorical data
Performing Univariate analysis on categorical data

### a. Countplot
It will give the frequency count of every category in col

```python
sns.countplot(x=df['Survived'])
```

```python
sns.countplot(df['Pclass'])
```

```python
sns.countplot(df['Sex'])
```

```python
sns.countplot(df['Embarked'])
```

### b. Pie chart
It will give the frequency count in terms of percentage of every category in a col

```python
plt.pie(df['Survived'].value_counts(), labels = df['Survived'].unique(), autopct='%.2f', data=df)
```

## 2. Numerical Data

### a. Histogram
It will give you the distribution of data

```python
# Min age is nearby 0 and max age is nearby 80 we see that travellers age b/w 18 to 45 are more
plt.hist(df['Age'])
```

### b. Distplot
Advance of histogram

With the help of distplot we also see that which skeweness is present in dataset Normal distribution, Left skewed, right skewed.

```python
# Line is called as KDE(Kernel Density Estimation)
# It will define you the probability of present number in data
# If you pick randomaly any one pessenger then Probability of his Age 40 is 1.5%
# If you pick randomaly any one pessenger then Probability of his Age 60 is 0.5%
sns.distplot(df['Age'])
```

```python
# There is another way of finding skeweness
# 0 means Perfectly skewed(Normal distribution)
# +ve value --> Positively skewed towards right side
# -ve value --> Negatively skewed towards left side
df['Age'].skew()
```

### c. Boxplot
It help to visualize the five-number summary

It will generally use when noise and outliers is present in dataset

```python
sns.boxplot(df['Age'])
```

# Bivariate and multivariate Analysis

```python
tips = sns.load_dataset('tips')
tips.sample(5)
```

```python
titanic = pd.read_csv(r'E:\CETPA\DA\titanic_data.csv')
titanic.sample(5)
```

```python
flights = sns.load_dataset('flights')
flights.sample(5)
```

```python
iris = sns.load_dataset('iris')
iris.sample(5)
```

## 1. Scatterplot (Numerical-Numerical)

```python
# Bivariate analysis
# There is a linear relationship b/w total_bill and tip if total_bill increases then tip is also inceases
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
# Multivariate analysis
# 5 relationships --> Relationships b/w two quantities, gender of both quantities, smoker, how many peoples come with one family
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2. Barplot (Numerical - Categorical)

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Fare'],hue=titanic['Sex'])
```

## 3. Boxplot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4. distplot (Numerical - Categorical)

```python
# Bivariate analysis
# Age of people who are not survived --> Blue color line
# Age of people who are survived --> Orange color line

# If age is b/w 0-20 then there is a more chance to get survived than unsurvived as we can see in graph
# If age is b/w 20-50 then there is a more chance to get unsurvived as we can see in graph
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

## 5. Heatmap (Categorical - Categorical)

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
# We got information like as --> In 3-class unsurvived people is more
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# Percentage of survived people in every class
# In Pclass-1 46% people being survived
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
# Visualize upper analysis
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
# Female are many more survived than male
(titanic.groupby('Sex').mean()['Survived']*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)
With the help of clustermap we see the closeness between values

```python
# Those who are not travel with SibSp(i.e, 0) they are more unsurvived(195)
# Those who travel with 1 SibSp then 56 were unsurvived and 54 were survived
pd.crosstab(titanic['SibSp'],titanic['Survived'])
```

```python
# Visualization of Upper o/p
# Those who travel with 5 or 8 SibSp then there survival rate are equal e.g, We see in chart both color are same
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

## 7. Pairplot
If you have multiple numerical col and you want to plot chart between each numerical col

It will automatically detect every numerical col in dataset and plot the scatterplot b/w them

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)
It is a special kind of scatter plot

You use lineplot when in x axis there is a time based no(Year,month,date,Time etc.)

```python
flights.sample(2)
```

```python
# Total passengers a/to each year
new = flights.groupby('year').sum().reset_index()
new
```

```python
# If we see the chart then we got Airlines linearly grow
sns.lineplot(new['year'],new['passengers'])
```

#### Pivot Table
if you have a large dataset and you need to analyze or summarize it in a flexible and dynamic way

```python
# How many passengers travlling in every year month wise
flights.pivot_table(values='passengers',index='month',columns='year')
```

```python
# Visualization of upper o/p
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```

```python
# Those months who behave equal these are in same color
# e.g, Jan, Feb of 1949
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
